### Описание задачи

Глава муниципалитета получает сведения о пуле согласованныз проектов по мощности электроснабжения.

Также он получает данные о резереве мощности на текущий момент

### Импорты

```python
import pandas as pd
import random
import numpy as np
import matplotlib.pyplot as plt
```

В виде таблице ниже управляющему приходит пул согласованным к будущему подключению мощностей.
- `Date of project approval` - дата согласования проекта по подключению к объекту требуемой мощности
- `Subscriber` - Абонент, тип
- `Requested Power` - согласованная мощность, кВт
- `District` - район, номер муниципалитета


```python


data = []
columns = ['Date of project approval', 'Subscriber', 'Requested Power', 'District']

for district in range(1, 11):
    data.append(['11.01.23', 'Торговый центр', random.randint(300, 700), district])
    data.append(['11.01.23', 'Школа', random.randint(200, 500), district])
    data.append(['11.01.23', 'Жилой комплекс', random.randint(100, 150), district])
    data.append(['11.01.23', 'Коттеджный поселок', random.randint(70, 200), district])

df_pool = pd.DataFrame(data, columns=columns)
display(df_pool)
```

Также на входе данные о количестве свободных мощностей по районам на текущую дату 

- `district` - номер района
- `Power,reserve` - свободная мощность

```python
data = [
    [1, 4000],
    [2, 5000],
    [3, 15000],
    [4, 500],
    [5, 3000],
    [6, 6000],
    [7, 7000],
    [8, 2000],
    [9, 500],
    [10, 10000]
]

columns = ['District', 'Power,reserve']

df_vacant_power = pd.DataFrame(data, columns=columns)
display(df_vacant_power)
```

Далее мы суммируем запрашиваемуюю мощнсоть по каждому району из таблицы df_pool

```python


# Your existing code to create df_pool goes here...

# Calculate the sum of power for each district
df_sum_power = df_pool.groupby('District')['Requested Power'].sum().reset_index()

display(df_sum_power)
```

```python
import pandas as pd

# Your existing code to create df_pool and df_sum_power goes here...

# Merge df_sum_power with the 'Power,reserve' DataFrame
df_merged = df_sum_power.merge(df_vacant_power, on='District', how='left')

# Calculate the difference between power reserve and requested power
df_merged['Difference'] = df_merged['Power,reserve'] - df_merged['Requested Power']

display(df_merged)
```

```python
# Your existing code to create df_merged goes here...

# Create a style function to highlight cells within a specific range in yellow, and others in red
def highlight_range(val):
    if 0 <= val <= 1000:
        background_color = 'yellow'
    elif val <0:
        background_color = 'red'

    else:
        background_color = 'black'
    return f'background-color: {background_color}'

# Create a styler object for the DataFrame and apply the style function
styled_df = df_merged.style.applymap(highlight_range, subset=['Difference'])

# Display the styled DataFrame
styled_df
```

После получения этих данных, у админстратора несколкьл сценариев:
 - реконструкция подстанции (установка трансформатора большей мощности)
 - строительство новой подстанции (Блочной БКТП для города, столбовой СТП или матчовой для коттеджных поселков и других маломощных объектов)
 - в случае еслм мощности не хватает на силовой посдтанции 35/10 кВ, то требуется реконструкция уже этой подстанции
 - вслучае особо крупных объектов, например завода, строительсвто собственной подстанции для эого завода 110/10 кВ например
